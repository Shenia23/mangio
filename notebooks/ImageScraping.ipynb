```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
recipes = pd.read_csv('../data/filtered_recipes_100.csv',sep='|')
```

```python
recipes.head()
```

```python
def getImgSrc(url):
    soup = BeautifulSoup(requests.get(url).text)
    try:
        img_div = soup.find("div", {"class": "imagen lupa"}).find('img')
    except AttributeError:
        print('Image not available in url: ', url )
        return np.nan
    return img_div['src']
```

```python
recipes.image_src = recipes.Link_receta.apply(getImgSrc)
```
